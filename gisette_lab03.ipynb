```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
%matplotlib inline

gisette_train = pd.read_csv('gisette_train.data', delim_whitespace=True)
label_df = pd.read_csv('gisette_train.labels', delim_whitespace=True).iloc[:,0]
```

```python
from sklearn.preprocessing import StandardScaler
gisette_train_x = StandardScaler().fit_transform(gisette_train)
```

```python
from sklearn.decomposition import PCA
pca_breast = PCA(n_components=2)
train_set_pca = pca_breast.fit_transform(gisette_train_x)
principal_breast_df_train = pd.DataFrame(data=train_set_pca, columns=['PC1', 'PC2'])
```

```python
pca_breast.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
plt.figure()
plt.xlabel('PC1')
plt.ylabel('PC2')
targets = (1, -1)
colors = ('b', 'r')
for target, color in zip(targets, colors):
    indicesToKeep = label_df == target
    plt.scatter(principal_breast_df_train.loc[indicesToKeep, 'PC1'],
               principal_breast_df_train.loc[indicesToKeep, 'PC2'],
               c = color)
```

```python
pca = PCA().fit(gisette_train_x)
evr = pca.explained_variance_ratio_
print('\n'.join('{:2d}: {:0.4f}'.format(*k) for k in enumerate(evr)))
```

```python
import numpy as np

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
```
